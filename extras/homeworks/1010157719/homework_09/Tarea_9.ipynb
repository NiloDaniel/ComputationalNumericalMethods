Solve the same equations using scipy.integrate.odeint and scipy.integrate.solve_ivp:
$$\frac{dx}{dt} = σ(y-x)$$

$$\frac{dy}{dt} = x(ρ-z)- y$$

$$\frac{dz}{dt} = xy - βz$$

for the parameters:

- $\sigma = 10$
- $ρ = 28$
- $\beta = 8/3$ 

lets set intial conditions:

- x(0)=1.0
- y(0)=1.0
- z(0)=1.0

Compare the results. Seeing anything strange?
 
 
 


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
```

```python
y_0 = [1.0,1.0,1.0] #Condiciones iniciales
sigma = 10.0 #parámetros
rho = 28.0
beta = 8/3

def f(t,y): #Funcion diferencial
  dx = sigma * (y[1] - y[0])
  dy = y[0] * (rho - y[2]) - y[1]
  dz = y[0] * y[1] - beta*y[2]
  return [dx,dy,dz]
```

```python
t = np.arange(0.0, 40.0, 0.01) #Rango a iterar en odeint
t_span=(0,40.0) #Rango a iterar en solve_ivp

soluciones = odeint(f, y_0, t, tfirst=True)
soluciones2 = solve_ivp(f, t_span, y_0)
```

```python
fig,ax=plt.subplots(figsize=(8, 6), dpi=80)
ax.plot(soluciones[:,0], soluciones[:,2], color = 'red')
ax.set_ylabel('Z ')
ax.set_xlabel('X')
ax.set_title('Gráfica Z vs X (Odeint)')
ax.legend()
ax.grid()
```

```python
fig,ax=plt.subplots(figsize=(8, 6), dpi=80)
ax.plot(soluciones2.y[0, :], soluciones2.y[2, :], color = 'blue')
ax.set_ylabel('Z ')
ax.set_xlabel('X')
ax.set_title('Gráfica Z vs X (Solve_ivp)')
ax.legend()
ax.grid()
```
