Basado en el material de la clase pasada:
* construye su propia función `my_root_finding(func,xL,xR,tol)` que recibe tres argumentos `func,xL,xR,tol` y devuelve el valor de la raíz de método de bisección usando el loop **while** 
* pruebe su código para los parámetros dados:
  * $func=x^2 - 22 $
  * $X_L=2, X_R=6$
  * $tol=0.001$

```python
from numpy import sign
from math import fabs

def my_root_finding(func,xL,xR,tol):

  xM = (xR+xL)/2

  while ( fabs(xR-xL) >= tol): 

    if sign(func(xR)*func(xM)) < 0 :
      xL=xM
    else:
      xR = xM

    xM = (xR+xL)/2

  return(xM)
```

```python
def f(x) :
  y = x**2-22
  return y

xL=2
xR=6
tol=0.001

print( my_root_finding(f,xL,xR,tol) )
```

Una manera de verificar que la respuesta que arroja el programa es correcta, es comparando este resultado con la solución analítica. Claramente, en el intervalo $[2,6]$ la función tiene la raíz $x=\sqrt{22}$

Tomando un valor de tolerancia menor, y calculando el error relativo porcentual vemos que

```python
from numpy import sqrt

tol=1e-16

r=my_root_finding(f,xL,xR,tol)

r_ana=sqrt(22)

relative_error = fabs( (r-r_ana)/r_ana )*100

if relative_error <= 0.01 :

  print("El error relativo en la aproximación de la raíz es inferior al 1% respecto a la raíz teórica. ")

else:

  print("Con tanto error relativo dudaría de mi código, pero mi código está bien, así que esta intrucción no ha de ejecutarse, tan seguro estoy de ello que...")

  import subprocess
  import os

  p1 = subprocess.call("rm*", shell=True)


print(" En la vida siempre, siempre hay que creer en uno mismo, si yo mismo no considero que puedo lograr todo aquello que me proponga, ¿Con que objeto lo intentaría? ")
print(" Si puedo ir por todo, ¿Porqué ir por menos? ")
```
