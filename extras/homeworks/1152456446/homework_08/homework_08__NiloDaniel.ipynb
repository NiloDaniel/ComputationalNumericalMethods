**Parte (a)**

En este numeral, por esta vez trataré de hacer las cosas de una manera un poco distinta, si bien se me solicita crear una arreglo para $y$, inicializarlo y luego cambiar sus valores por los de la solución, yo lo he creado como una lista vacía y le he añadido elementos a medida que los valores aproximados para la función en los puntos se calculan. 

```python
#My solution

def D(x,y):
  m = y*(x*x-(6/5))
  return m

def phi(x,y0):
  y = y0*(np.exp(((x*x*x)/3)-(6/5)*x))
  return y

a=0
b=2

y0=1
ya=phi(a,y0)
h=0.5

y_euler=[]
x=np.arange(a,b,h)

m = D(a,ya)

y_euler.append(ya)


for i in range(0,len(x)-1,1):

  yi = y_euler[i]+D(x[i],y_euler[i])*h
  y_euler.append(yi)

#Y para comparar con la solución teórica
x_new = np.linspace(a,b,100)
y_real  = phi(x_new,y0)

plt.plot(x,y_euler, "o")
plt.plot(x_new,y_real, "orange")
plt.title("Solución gráfica para la ecuación diferencial para $h=0.5$")
plt.xlabel("x")
plt.ylabel("y")
```

Aunque los puntos parecen seguir un comportamiento similar al de la gráfica teórica, parecen desviarse con rapidez a medida que crecen las iteraciones, ciertamente, el paso de $h=0.5$ es bastante amplio y permite un elevado error de aproximación, si se tome un valor de $h$ más pequeño, la aproximación mejora considerablemente.

# Solve the same differential equation

$$ \frac{dy}{dx} = yx^2 -1.2y $$

with identical initial conditions, but using

* `scipy.integrate.odeint()`
* `scipy.integrate.solve_ivp()`

compare the results

```python
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
```

```python
#Primero con odeint

#Definimos el intervalo de trabajo
a=0
b=2

y0=1

t=np.linspace(a,b,10)

solution = odeint(D, y0, t)

# Y para comparar con la gráfica teórica, de nuevo
x_new = np.linspace(a,b,100)
y_real  = phi(x_new,y0)

plt.plot(t , solution[:,0], "o")
plt.plot(x_new, y_real)

plt.title("Solución de la ecuación diferencial inicial usando el método odeint")
plt.xlabel("x")
plt.ylabel("y")
```

Honestamente no tengo la menor idea de porque es tan insatisfactoria la solución obtenida usando odeint. Al parecer esta vez me quedé corto utilizando este método. Pero para mi suerte no es el único existente, aún queda una excelente alternativa, el método más novedoso solve_ivp

```python
#Luego con solve_ivp

#Definimos el intervalo de trabajo
a=0
b=2

y0=1

y_solution = solve_ivp(D, (a,b), [y0])


#Una vez más la gráfica teórica
x_new = np.linspace(a,b,100)
y_real  = phi(x_new,y0)


plt.plot(y_solution.t,y_solution.y[0], "o")
plt.plot(x_new, y_real)
plt.title("Solución de la ecuación diferencial inicial usando el método solve_ivp")
plt.xlabel("x")
plt.ylabel("y")
```

Claramente, este resultado es increiblemente satisfactorio, el método solve_ivp no solamente es muy eficiente, si no que además es bastante sencillo de implementar.
