## Last homework
Solve the same equations using `scipy.integrate.odeint` and `scipy.integrate.solve_ivp`

$$
\begin{align*} 
\frac{dx}{dt}&=\sigma (y-x) \\
\frac{dy}{dt}&=x(\rho-z) -y \\
\frac{dz}{dt}&=xy-\beta z 
\end{align*} 
$$

for the parameters:

* $\sigma =10$
* $\rho=28$
* $\beta=8/3$

lets set initial conditions:

* x(0)=1.0
* y(0)=1.0
* z(0)=1.0

Compare the results. Seeing anything strange?

That's my notation. 

$R$ is a vector made with the three variables

$ R = \begin{bmatrix}
x \\
y \\
z
\end{bmatrix} $

$ DR $ (derivs) has another vector made with the equations members in the right side.

The differential equation could be expresed thus

$ R = DR $

```python
#I need some powerful toys in my bag

from scipy.integrate import odeint
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import numpy as np
```

```python
sigma=10
ro=28
beta=8/3

def Derivs(R,t):

  x=R[0]
  y=R[1]
  z=R[2]

#Se inicializa el vector de derivadas para poder almacenar los lados derechos de la ecuaci√≥n
  DR=[0,0,0]

  DR[0]=sigma*(y-x)
  DR[1]=x*(ro-z)-y
  DR[2]=x*y-beta*z

  return DR
```

```python
#I can take an interval from 100 unitys of t with 100 points
t0=0
tf=100
n=10000

t=np.linspace(0,100,n)

#It's clear that

R0=(1,1,1)
```

```python
Solution = odeint(Derivs, R0, t)

#print(Solution)

plt.plot(t, Solution[:,0], color="blue", label="x(t)")
plt.plot(t, Solution[:,1], color="red", label="y(t)")
plt.plot(t, Solution[:,2], color="yellow", label="z(t)")

plt.legend(loc="upper left")
```

```python
from mpl_toolkits import mplot3d

%matplotlib inline
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot3D(Solution[:,0], Solution[:,1], Solution[:,2], 'blue')
```
